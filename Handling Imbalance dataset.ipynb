### How to handled imbalance dataset

#### Separating the legit and fraudulent transactions

```python
legit = credit_data[credit_data.Class == 0]
fraud = credit_data[credit_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

### We are using under sampling method 

Building a sample dataset containing similiar distribution of Legit and Fraudulent transaction

Number of Fraudulent transactions  -- > 492

```python
legit_sample = legit.sample(n=492)
```

```python
print(legit_sample.shape)
```

#### Concatenate the two dataframes

```python
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
```
